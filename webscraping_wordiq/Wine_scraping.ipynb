```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
```

```python
link = 'http://www.snooth.com/talk/topic/whatcha-drinking-tonight/'

req = requests.get(link)

text = BeautifulSoup(req.text, "html.parser")

stats = text.findAll('div', {'class': 'talkpost'})
```

```python
posts = []
for i in range(2,50):
    link = 'http://www.snooth.com/talk/topic/whatcha-drinking-tonight/' + str(i)
    req = requests.get(link)
    text = BeautifulSoup(req.text, "html.parser")
    stats = text.findAll('div', {'class': 'talkpost'})
    posts.extend(stats)
```

```python
posts[0].p
```

```python
final = []
for i in range(len(posts)):
    string = str(posts[i].p)
    string = re.sub('\xc2\xa0',' ', string)
    string = re.sub('<[^>]+>', '', string)
    string = re.sub('\xc3\xa9','e',string)
    string = string[2:]
    string.replace("\\",'')
    final.append(string)
```
